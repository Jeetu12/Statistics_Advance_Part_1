1. **What is a random variable in probability theory?**  

- Random Variable in Probability Theory:

  - A random variable is a function that assigns a numerical value to each possible outcome of a random experiment. It helps quantify uncertainty in a mathematical way.

---

2. **What are the types of random variables?**  

- There are two types of random variables:

   - **Discrete Random Variables:** Take countable values (e.g., number of heads in coin flips).  

   - **Continuous Random Variables:** Take an infinite number of values within a range (e.g., height, temperature).

---

3. **What is the difference between discrete and continuous distributions?**  

- **Discrete distributions** describe random variables with distinct, separate values.  
   
- **Continuous distributions** describe random variables that take any value within a given range.

---

4. **What are probability distribution functions (PDF)?**  
   
- Probability Distribution Functions (PDFs):

 - A Probability Distribution Function (PDF) describes the likelihood of a random variable taking specific values. PDFs can be classified into two types based on whether the random variable is discrete or continuous.

5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**  
   
- **PDF** gives the probability of a specific value occurring.  
   
- **CDF** gives the probability that a random variable is less than or equal to a given value.

7. **What are the key properties of a Bernoulli distribution?**  
   
- The **Bernoulli distribution** is a discrete probability distribution that models a single experiment with two possible outcomes: success (1) and failure (0). It is one of the simplest probability distributions and forms the basis for the Binomial distribution.  

- #### **Key Properties of Bernoulli Distribution:**  

1. **Two Possible Outcomes** – The random variable X can take only two values:  
   
- X = 1 (Success) with probability p  

- X = 0 (Failure) with probability 1 - p

2. **Probability Mass Function (PMF)** – The probability of X taking a specific value is given by:  

- P(X = x) = p^x (1 - p)^{(1 - x)}, quad x in {0,1}
   where p is the probability of success.

3. **Mean (Expected Value)** – The expected value of X is:  
   
- E[X] = p

- This represents the long-run average outcome.

4. **Variance** – The variance of X is:  

- Var(X) = p(1 - p)

- It measures the spread of outcomes.

5. **Standard Deviation** – Given by:  
   
- sigma = sqrt{p(1 - p)}

6. **Skewness** – The skewness of the Bernoulli distribution is:  
   
- {1 - 2p}/{sqrt{p(1 - p)}}

   - If p = 0.5, it is symmetric.  
   - If p < 0.5, it is skewed right.  
   - If p > 0.5, it is skewed left.

7. **Kurtosis** – The excess kurtosis of a Bernoulli distribution is: {6p(1 - p)}/{p(1 - p)} = {6}/{p(1 - p)} - 3

- It depends on p and can be high for extreme values of p.

8. **Memoryless Property** – Unlike the geometric and exponential distributions, the Bernoulli distribution **does not** have the memoryless property.

9. **Special Case of Binomial Distribution** – A Bernoulli trial is a **Binomial distribution** with a single trial (\( n = 1 \)):  X ~ {Binomial}(n=1, p)

10. **Applications** – The Bernoulli distribution is widely used in:  
   - **Machine learning:** Binary classification (e.g., spam detection: spam or not spam).  

   - **Finance:** Modeling success/failure of investments.  

   - **A/B Testing:** Checking the success of a new feature.  

   - **AI & Deep Learning:** Activation functions (e.g., dropout in neural networks).

12. **What is the standard normal distribution, and why is it important?**  
    
- #### **Standard Normal Distribution:**  

 - The **standard normal distribution** is a special case of the **normal distribution**, where the **mean (mu) is 0** and the **standard deviation (sigma) is 1**. It is denoted as:
Z ~ N(0,1)

- #### **Importance of the Standard Normal Distribution:**  

1. **Standardization & Z-Scores** –  
   
   - Any normal distribution **N(mu, sigma^2)** can be transformed into a standard normal distribution using:

     Z = (X - mu)/sigma

   - This transformation (Z-score) helps compare values from different normal distributions.

2. **Probability Calculations** –  
   
   - The standard normal table (Z-table) is used to find probabilities and percentiles.

3. **Central Limit Theorem (CLT)** –  
   
   - The CLT states that, for large sample sizes, the sampling distribution of the mean approaches a normal distribution, making the standard normal critical in statistical inference.

4. **Hypothesis Testing & Confidence Intervals** –  
   
   - Used to compute **p-values** and **confidence intervals**
   (e.g., 95% confidence interval corresponds to Z = 1.96, -1.96.)

5. **Machine Learning & Data Science** –  
  
   - Many ML models assume normally distributed data.
  
   - Feature scaling techniques like **standardization** (zero mean, unit variance) improve model performance.

6. **Signal Processing & AI** –  
   
   - Noise in real-world signals is often modeled as Gaussian (normal) noise.

17. **What are point estimates and interval estimates in statistics?**

18. **What is the significance of confidence intervals in statistical analysis?**  
  
- Confidence intervals (CIs) are essential in statistical analysis because they provide a range of values within which a population parameter (such as a mean or proportion) is likely to fall, with a specified level of confidence (e.g., 95% or 99%). Here’s why they are significant:

1. **Measure of Uncertainty** – CIs help quantify the uncertainty associated with sample estimates, providing a more informative result than a single point estimate.

2. **Inference About Population** – Since data is usually collected from a sample rather than the entire population, confidence intervals help infer the true population parameter.

3. **Decision-Making Tool** – In fields like business, medicine, and engineering, CIs guide decision-making by indicating the reliability of estimates (e.g., whether a new drug improves recovery rates compared to an existing treatment).

4. **Hypothesis Testing** – CIs are closely related to hypothesis testing. If a confidence interval for a parameter does not include a hypothesized value (e.g., zero in the case of a difference in means), it suggests statistical significance.

5. **Comparison of Groups** – CIs allow comparisons between groups by checking whether confidence intervals overlap. Non-overlapping intervals often indicate significant differences.

6. **Robustness Against Sample Variability** – Unlike point estimates, CIs take sample variability into account, making them more informative and reliable in real-world applications.

- ### Example:

 - If a 95% confidence interval for the mean height of students is (165 cm, 175 cm), it suggests that we are 95% confident that the true mean height of all students falls within this range.
